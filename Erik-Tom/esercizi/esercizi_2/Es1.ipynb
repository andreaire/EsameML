```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris
import numpy as np
from pathlib import Path

df=pd.DataFrame({
    'ID': [1,2,3,4,5],
    'Importo': [100.5,2500.0,80.0,None,1200.0],
    'Categoria' : ['Retail','Online','Retail','Online','Online'],
    'Target': [0,1,0,0,1]
})


print(f"Dimensioni del dataset: {df.shape}")
print(f"\nColonne: {list(df.columns)}")

print("\n")
df.info()
df.isna().sum()

print("\n")
df.describe()


```

## SLICING


```python
slicing = df.iloc[0:3, df.columns.get_indexer(['Importo','Categoria'])]
# df.iloc[0:3,[1,2]] -- al posto di prendere le colonne singolarmente le prende utilizzando gli indici
print(slicing)
```

## FILTRO LOGICO

```python
importo=df[df['Importo']>500]
print(importo)
```

## FILTRO ML

```python
frodi = df[(df['Target'] == 1) & (df['Importo'].notna())]
print(frodi)
```

## CONCAT

```python
df2=pd.DataFrame({
    'ID': [6],
    'Importo': [50.0],
    'Categoria' : [None],
    'Target': [0]
})

dfconcatenato=pd.concat([df,df2], axis=0)
print(dfconcatenato)
```
